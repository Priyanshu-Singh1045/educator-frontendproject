## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
merged_df['description'].fillna('No description available', inplace=True)
merged_df['age_certification'].fillna('Not Rated', inplace=True)
merged_df['seasons'].fillna(0, inplace=True)
merged_df['character'].fillna('Unknown', inplace=True)
merged_df['name'].fillna('Unknown', inplace=True)
merged_df['role'].fillna('Unknown', inplace=True)
merged_df['imdb_id'].fillna('No IMDb ID', inplace=True)


merged_df['imdb_score'].fillna(merged_df['imdb_score'].median(),inplace=True)
merged_df['imdb_votes'].fillna(merged_df['imdb_votes'].mean(), inplace=True)
merged_df['tmdb_popularity'].fillna(merged_df['tmdb_popularity'].median(), inplace=True)
merged_df['tmdb_score'].fillna(merged_df['tmdb_score'].median(), inplace=True)

merged_df.dropna(subset=['person_id'], inplace=True)
````

### What all manipulations have you done and insights you found?

1. **Filling Text Fields with Placeholders description** Why: A small number of titles are missing descriptions. Action: Fill with ‘No description’ to avoid leaving nulls in visual summaries or dashboards.

**age_certification** - Why: Over half the titles are missing age certifications.Action: Replace with ‘Not Rated’, which is commonly used for unrated content.

**seasons** - Why: Most titles are movies and don’t have seasons.Action: Fill with 0 to denote it’s not a TV series.

**character** - Why: Some actors don’t have a listed character name.Action: Fill with ‘Unknown’ for completeness in cast listings.

**name, role** - Why: Rare but important to identify people in credits.Action: Use ‘Unknown’ as a neutral placeholder.

**imdb_id** - Why: imdb_id is a unique identifier; missing values can’t be inferred or used in joins.Action: Replace missing entries with ‘missing_id’ to ensure consistency and prevent null-related issues.

2. **Filling Numerical Columns with Medianimdb_score, imdb_votes** - Why: IMDb scores and votes are important indicators of popularity and quality.Action: Median is used to avoid skewing from outliers.

**tmdb_popularity, tmdb_score** - Why: Similar to IMDb values—important for popularity analysis.Action: Median provides a balanced fill for small gaps.

3. **Dropping Critical Null Rowsperson_id** - Why: person_id is a unique identifier for individuals in the cast/crew.Action: If it’s missing, the row lacks fundamental data and is dropped.
